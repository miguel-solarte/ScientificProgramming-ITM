## 1. Data loadiong, preprocessing and initial vizulization.

The dataset of QAM constellations was generated using the MATLAB Communications Toolbox with the purpose of obtaining a dataset containing constellations such as 16-QAM, 32-QAM, and 64-QAM. Non-linear noise was added to each constellation to emulate the conditions caused by light passing through a long optical channel. Additionally, each constellation was varied by adjusting inherent features like signal-to-noise ratio (SNR) with a range from 5 dB to 20 dB, and frequency offset, with values of -90, -100, and -110 dB/Hz at 1 MHz, for a transmission frequency of 10 GHz.

### Proper data loading and handling

If you need to load the dataset, it is necessary to write the path of the file with the .csv extension that you choose. Since the ideal constellation saves important information, you should load a dataframe (```pd.read_csv(path, dtype={1: str})```). With the dataframe of the ideal constellations, you will be able to obtain measures like the bit error rate (BER). For the dataset with nonlinear noise in the constellation, you should load it as a NumPy array using the ```numpy.genfromtxt(path, delimiter=',')``` method, where the first row represents the In-phase component, the second row the Quadrature component, and the third row contains the labels. ## 2. Python fundamentals and OOP for data processing

The following code generates ideal constellations. 

````python
from itertools import product
````

````python
class Constellation_mQAM():
    def __init__(self):
        self.labels_decimal, self.labels_bin, self.x, self.y = self.construct_method()

    def construct_method(self):
        raise NotImplementedError('no rule in class %s' %self.__class__.__name__)
    
    def _binary_to_gray(self, binario, bits):
        num = int(binario, 2)
        gray = num ^ (num >> 1)
        return f"{gray:0{bits}b}"

    def generate_constellation_nQAM(self, path_mQAM):

        data = {'labels_decimal': self.labels_decimal,
                'labels_bin': self.labels_bin,
                'x': self.x,
                'y': self.y}
    
        df = pd.DataFrame(data)
        df['labels_bin'] = df['labels_bin'].astype(str)
        df.to_csv(path_mQAM, index =  False)
````

````python
class Constellation_16QAM(Constellation_mQAM):
    def construct_method(self):

        y = []

        flag = -1
        
        binary_to_gray = self._binary_to_gray

        labels_bin = [binary_to_gray(''.join(seq), 4) for seq in product('01', repeat=4)]
        labels_decimal = [int(bin_str, 2) for bin_str in labels_bin]

        for _ in range(4):
            for i in [-3,-1,1,3]:
                y.append(i*flag)
            flag = flag * -1


        x = [i for i in [-3,-1,1,3] for _ in range(4)]

        return labels_decimal, labels_decimal, x, y
````

````python
class Constellation_32QAM(Constellation_mQAM):
    def construct_method(self):

        binary_to_gray = self._binary_to_gray
        
        labels_bin = [binary_to_gray(''.join(seq), 5) for seq in product('01', repeat=5)]
        labels_decimal = [int(bin_str, 2) for bin_str in labels_bin]

        x = [-3, -1, -1, -3, -5, -5, -5, -5, -3, -3, -3, -3, -1, -1, -1, -1, 1, 1, 1, 1, 3, 3, 3, 3, 5, 5, 5, 5, 3, 1, 1, 3]
        y = [ 5,  5, -5, -5, -3, -1,  1,  3,  3,  1, -1, -3, -3, -1,  1, 3, 3, 1, -1, -3, -3, -1, 1, 3, 3, 1, -1, -3, -5, -5, 5, 5]

        return labels_decimal, labels_decimal, x, y
````

````python
class Constellation_64QAM(Constellation_mQAM):
    def construct_method(self):

        y = []

        flag = -1
        
        binary_to_gray = self._binary_to_gray

        labels_bin = [binary_to_gray(''.join(seq), 6) for seq in product('01', repeat=6)]
        labels_decimal = [int(bin_str, 2) for bin_str in labels_bin]

        y = []

        for _ in range(8):
            for i in [-7,-5,-3,-1,1,3,5,7]:
                y.append(i*flag)
            flag = flag * -1


        x = [i for i in [-7,-5,-3,-1,1,3,5,7] for _ in range(8)]

        return labels_decimal, labels_decimal, x, y
````

The code consists of four classes: the first one is the Parent class (`Constellation_mQAM`), and the other three are Child classes (`Constellation_16QAM(Constellation_mQAM)`, `Constellation_32QAM(Constellation_mQAM)`, and `Constellation_64QAM(Constellation_mQAM)`). The Child classes inherit the `construct_method` from the Parent class. Additionally, the Child classes call an encapsulated method from the Parent class, `_binary_to_gray.`

````python
path_qam16 = "../ideal_constQAM_generates/ideal_constellation_16QAM.csv"
QAM16 =  Constellation_16QAM()

path_qam32 = "../ideal_constQAM_generates/ideal_constellation_32QAM.csv"
QAM32 =  Constellation_32QAM()

path_qam64 = "../ideal_constQAM_generates/ideal_constellation_64QAM.csv"
QAM64 =  Constellation_64QAM()
````

````python
QAM16.generate_constellation_nQAM(path_qam16)
QAM32.generate_constellation_nQAM(path_qam32)
QAM64.generate_constellation_nQAM(path_qam64)
````

## 3. Numerical derivation or integration of data

This part discusses the derivation of the image generated for the mQAM constellation. The derivative was implemented for edge detection. Before performing edge detection, it is necessary to apply an erosion process to connect the points. If the erosion process is not applied, you will obtain the same image as if only the edge detection process was applied.

````python
data_16QAM = genfromtxt("./Data_constellations_m-QAM/Constellation_16QAM/sintetic_data_16QAM_level90/data_16QAM_SNR_15_level90.csv", delimiter = ',')
features_data_16QAM = data_16QAM[:,:2]
````

````python
import cv2
import numpy as np
````

````python
def convolution(image, kernel):
    size_image_x = image.shape[1]
    size_image_y = image.shape[0]

    size_kernel_x = kernel.shape[1]
    size_kernel_y = kernel.shape[0]

    conv = np.zeros((size_image_y, size_image_x))

    for i in range(size_image_y - size_kernel_y):
        for j in range(size_image_x - size_kernel_x):
            conv[i,j] = np.sum(image[i:i+size_kernel_y, j:j+size_kernel_x] * kernel)

    return conv
````

````python
image = cv2.imread("../images/data_16QAM.png", cv2.IMREAD_GRAYSCALE)

kernel_binary = np.ones((5,5))

eroded_image = cv2.erode(image, kernel_binary, iterations=1)
````

````python
plt.figure()
plt.imshow(eroded_image, cmap='gray')
plt.xticks([]), plt.yticks([])
plt.show()
````

````python
dev_kernel = np.array([[-1, 0, 1],
                       [-1, 0, 1],
                       [-1, 0, 1]])

edge_image = convolution(eroded_image, dev_kernel) 
````

The previously performed process is necessary for analyzing the number of potential clusters over the edge-detected image. Knowing the number of clusters is essential when using arbitrary constellations to identify the type of constellation.

````python
plt.figure()
plt.imshow(np.clip(edge_image, 0, 255), cmap='gray')
plt.xticks([]), plt.yticks([])
plt.show()
````

## 4. Matrix operations

The algorithm called *On spectral clustering* was obtained from the paper: https://ai.stanford.edu/~ang/papers/nips01-spectral.pdf

````python
from sklearn.datasets import make_circles

import numpy as np
import scipy as sp
````

````python
def on_spectral_clustering(sigma, features_data):
    

    distances = sp.spatial.distance.cdist(features_data, features_data, metric='euclidean')

    A = np.exp(-np.power(distances,2) / (2 * np.power(sigma, 2)))
    np.fill_diagonal(A, 0)


    D_inv = np.diag(1 / np.sqrt(np.sum(A, axis = 1)))

    L = D_inv @ A @ D_inv


    _, eigenvectors = sp.sparse.linalg.eigsh(L, 2)

    X = np.copy(eigenvectors)

    X_normalized = np.sqrt(np.sum(np.power(X, 2),axis = 1)).reshape(len(X),1)

    Y = X / X_normalized

    return Y
````

````python
circles_features, _ = make_circles(n_samples = 1000, factor = 0.5, noise = 0.05, random_state = 42)
````

````python
space_eigenvect = on_spectral_clustering(0.1, circles_features)
````

The algorithm implemented in the function on_spectral_clustering was tested with synthetic data obtained using the `make_circles` method from Scikit-learn, instead of the mQAM constellation dataset, because the results obtained from the spectral clustering algorithm are more useful for explanation and plotting.

````python
fig, (axes1,axes2) = plt.subplots(nrows=1, ncols=2)
axes1.scatter(circles_features[:,0], circles_features[:,1], c = 'blue')
axes1.set_title("Points in the original space")

axes2.scatter(space_eigenvect[:,0], space_eigenvect[:,1], c = 'blue')
axes2.set_title("Points in the space of the eigenvectors")

plt.tight_layout()
plt.show()
````

The spectral clustering algorithm is divided into two parts. The first part is implemented with matrix operations to obtain a different representation of the data in the space generated by eigenvectors. The second part applies k-means to the data in this eigenvector space. If in this case, you apply k-means directly to the original data space, it cannot divide the data properly. However, if you apply k-means in the space generated by the eigenvectors, k-means can divide the data correctly.
