# 3. Data separability

To calculate and correctly implement appropriate performance metrics, we will first implement a machine learning algorithm called spectral clustering.

````python
from scipy.spatial.distance import cdist
from sklearn.cluster import SpectralClustering
````

The function `batches(data)` was create for two reason:

1. To reduce the number of symbols in each iteration of the spectral clustering algorithm. Due to its computational complexity, the spectral clustering algorithm has a complexity of $ O(n^{3})$ and Using all the data **in this case, 100,000 symbols** would be problematic and require extensive computational resources.

2. To transition from static data to streaming data. (Note: Streaming data and batch data are not the same and will be discussed further in Lab 3.)

````python
def batches(data):

    for datos_2048 in range(0, len(data), 2048):

    
        if datos_2048 + 2048 <= len(data):
            buffer_2048 = data[datos_2048: datos_2048 + 2048, :]

            yield buffer_2048
        
        else:
            buffer_2048 = data[datos_2048:, :]

            yield buffer_2048  
````

In practice, each batch changes over time and should be processed by the spectral clustering algorithm. Although spectral clustering produces appropriate clusters, it may present a challenge with label inconsistency, leading to incorrect demapping. This issue is illustrated in Figure 1a and Figure 1b, which show a switch in the labels assigned to two different batches. To address this, a `relabeling()` function was developed to achieve accurate demapping.

| ![First execution of the algorithm](../Figures/labels_iter_one.png) | ![Second execution of the algorithm](../Figures/labels_iter_two.png) |
|:--:|:--:|
| **Figure 1a**: First execution of the algorithm | **Figure 1a**: Second execution of the algorithm |

**Figure 1**: Label inconsistency.


````python
def relabeling(data, coor_constellation, labels_constellation, labels_model, num_cent):

    copy_labels_model = np.copy(labels_model)

    for i in range(num_cent):
        mean_labels = np.mean(data[labels_model == i], axis = 0).reshape(1,2)
        distances_mean_labels_to_cent = cdist(mean_labels, coor_constellation, metric='euclidean')
        min_distances_labels_to_cent = np.argmin(distances_mean_labels_to_cent)
        copy_labels_model[labels_model == i] = labels_constellation[min_distances_labels_to_cent]

    return copy_labels_model
````

The `generate_labels_osc()` function returns the labels of the clusters generated by the spectral clustering algorithm.

````python
def generate_labes_osc(data, gamma, num_cent):

    onspectral = SpectralClustering(n_clusters = num_cent, eigen_solver=None, n_components=None, 
                                random_state=None, n_init=10, gamma = gamma, affinity='rbf', 
                                n_neighbors=10, eigen_tol='auto', assign_labels='kmeans', 
                                degree=3, coef0=1, kernel_params=None, n_jobs=None, verbose=False).fit(data)

    labels_spectralclustering = onspectral.labels_

    return labels_spectralclustering
````

### Graphics BER vs SNR for contelations 16QAM 32QAM 64QAM

The Bit Error Rate indicates the number of error in a digital data transmission. BER is defined as:

$$
BER = \frac{\text{number erroneous bits}}{\text{total bits transmitted}}
$$


A low BER indicates a more accurate transmission, while a high BER indicates a greater number of errors, which could be due to interference, noise, or synchronization issues in the communication channel.

BER is an appropriate performance metric to evaluate the spectral clustering algorithm. However, since we used constellations with different SNRs (Signal-to-Noise Ratios), which is a measure that compares the power of a useful signal to the power of the background noise that accompanies it, we should create a graphic that shows how the BER varies with the signal power.


The `dec_to_bin()` function converts the decimal values of the labels returned by the spectral clustering algorithm into binary.

````python
def dec_to_bin(array_int, bits):

    list_bits = [np.binary_repr(x, width=bits) for x in array_int]
    n2 = ''.join(list_bits)
    n2 = ','.join(n2)

    return np.array(n2.split(','))
````

The stream_osc() function unites all the functions previously developed to return the values of the BER for each value of SNR.

Note: Each value of SNR is saved in a list called`values_SNR = [5,7,10,12,15,17,20]`

````python
def stream_osc(path_const, path_data, sigma, num_cent, bits):

    df_const_QAM = pd.read_csv(path_const, dtype={1: str})
    data_QAM = genfromtxt(path_data, delimiter=',')
    labels_data_QAM = data_QAM[:,-1].astype(int)

    
    features_QAM = data_QAM[:,:2]

    array_coor_QAM = np.array(df_const_QAM.iloc[:,2:])
    array_true_labels_QAM = np.array(df_const_QAM['labels_decimal'])

    list_labels_osc_QAM = []
    
    for array_data in batches(features_QAM):
    
        labels_osc = generate_labes_osc(array_data, sigma, num_cent)

        relabeled_osc = relabeling(array_data, array_coor_QAM, 
                                                   array_true_labels_QAM, labels_osc, num_cent)
    
        list_labels_osc_QAM.append(relabeled_osc.tolist())
    
    labels_osc_ = [j for i in list_labels_osc_QAM for j in i]
    labels_osc_ = np.array(labels_osc_).astype(int) 

    labels_osc_bin = dec_to_bin(labels_osc_, bits)
    labels_data_QAM = dec_to_bin(labels_data_QAM, bits)

    BER_osc = sum(labels_osc_bin != labels_data_QAM) / (len(labels_data_QAM) * bits)

    
    
    return BER_osc, labels_osc_
````

Figure 2 shows that for all QAM modulations, the BER values decrease as the SNR values increase. This is because a high SNR indicates that the signal is much stronger than the noise, generally leading to clearer and more accurate transmission. A low SNR suggests that the noise is comparable to the signal, which can lead to errors in communication or data processing.

For all QAM modulations, the spectral clustering algorithm performs poorly when the constellation level is -90 dB, but it shows better performance at levels -100 and -110. However, it is necessary to compare this with other algorithms to determine which is the best option for the demodulation task.
